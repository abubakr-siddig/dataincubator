```python
crimes_area = data.groupby([data[data['Charge Group Code']!=99]['Charge Group Code'],'Area ID'])['Report ID'].count().reset_index(name="count")
crimes_area.head()
```

```python
crimes_sum = []
perc = []
for code in crimes_area['Charge Group Code']:
    temp1 = crimes_area.groupby(crimes_area[crimes_area['Charge Group Code']==code]['Charge Group Code'])['count'].sum().reset_index(name="sum")
    temp2 = crimes[crimes['Charge Group Code']==code].reset_index()
    perc.append(temp2.at[0,'percentage'])
    crimes_sum.append(temp1.at[0,'sum'])
    
crimes_area['group_sum'] = crimes_sum
crimes_area['percentage_area'] = crimes_area['count']/crimes_area['group_sum']*100
crimes_area['percentage'] = perc
crimes_area['ratio'] = crimes_area['percentage_area']/crimes_area['percentage']
#print temp.at[0,'sum']

crimes_area
```
